# Simpletron - Um Computador Didático

O Simpletron é um computador didático fictício projetado para auxiliar no aprendizado e compreensão dos conceitos básicos da programação. Ele foi criado como uma ferramenta de ensino para introduzir os fundamentos da lógica de programação e familiarizar os estudantes com a arquitetura e o funcionamento de um computador.

## Características do Simpletron

O Simpletron criado possui as seguintes características:

Memória: O Simpletron possui uma memória com 100 células numeradas de 00 a 99. Cada célula pode armazenar uma palavra de instrução ou um valor de dados.

Acumulador: O Simpletron tem um acumulador (ACC) especial usado para armazenar temporariamente resultados de operações.

Instruções: O Simpletron utiliza uma linguagem de montagem simples composta por instruções numéricas de três dígitos. Cada instrução contém um código de operação e um endereço de operando.

Operações: O Simpletron é capaz de realizar operações de entrada e saída, operações aritméticas (soma, subtração, multiplicação, divisão) e transferências de controle (desvios condicionais e incondicionais).

Execução Passo a Passo: O Simpletron permite a execução passo a passo do programa, permitindo que os estudantes acompanhem o funcionamento interno do computador e observem como os dados e as instruções são processados.


## Utilizando o Simpletron

Para utilizar o Simpletron, deve-se escrever programas simples usando a linguagem de montagem do Simpletron, no meu caso foi escolhido a linguagem c++.Apos a escolha da linguagem agora basta inserir instruções e dados na memória do Simpletron e executar o programa para observar o resultado.

Os programas são compostos por uma sequência de instruções que são carregadas na memória e executadas pelo Simpletron. Os estudantes podem interagir com o Simpletron, inserindo valores de entrada e visualizando os resultados das operações.

Se o código de operação for "Branchneg" (41), o desvio ocorre se o valor do acumulador (ACC) for negativo.
Se o código de operação for "Branchzero" (42), o desvio ocorre se o valor do acumulador (ACC) for zero.
